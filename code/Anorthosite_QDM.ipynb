# Notebook for displaying QDM maps of the Beaver River anorthosite vs. Duluth Complex Anorthositic Series anorthosite

```python
import pmagpy.ipmag as ipmag
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import imageio
from PIL import Image
import matplotlib.cm as cm
%config InlineBackend.figure_format = 'retina'
```

## MS99033 IRM parallel to the long axis of the thin section (parallel to the y axis in the QDM field of view)

```python
MS99033_parallel_Bz = loadmat('../data/QDM/MS99033/Parallel_IRM/Bz_uc0.mat')
MS99033_parallel_Bz = MS99033_parallel_Bz['Bz'].reshape(300,374)

MS99033_paralleL_LED = Image.open('../data/QDM/MS99033/Parallel_IRM/LED.jpg').resize((374,300)).convert('LA')
```

```python
xticks_list = np.array([0, 50, 100, 150, 200, 250, 300, 350])
yticks_list = np.array([0, 50, 100, 150, 200, 250])
```

```python
vmin_QDM=-1e-5
vmax_QDM=1e-5

plt.figure(figsize = (37.4/3, 10))
# plt.imshow(MS99033_LED,alpha=0.6,cmap='Greys')
plt.imshow(np.flip(MS99033_parallel_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM, alpha = .6)
cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.0e')
cbar.ax.set_xlabel('Bz (T)', fontsize=12)

cbar.update_ticks()
plt.xticks(ticks = xticks_list, labels = xticks_list* 4)
plt.yticks(yticks_list,yticks_list* 4)
plt.savefig('../figure/MS99033_para_Bz_plot_with_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```
