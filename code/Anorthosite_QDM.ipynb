# Notebook for displaying QDM maps of the Beaver River anorthosite vs. Duluth Complex Anorthositic Series anorthosite

```python
import pmagpy.ipmag as ipmag
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import imageio
from PIL import Image
import matplotlib.cm as cm
%config InlineBackend.figure_format = 'retina'
```

## MS99033

```python
MS99033_Bz = loadmat('../data/QDM/MS99033/Bz_uc0.mat')
MS99033_Bz = MS99033_Bz['Bz'].reshape(300,374)

MS99033_LED = Image.open('../data/QDM/MS99033/LED.jpg').resize((374,300)).convert('LA')
```

```python
vmin_QDM=-1e-5
vmax_QDM=1e-5

plt.figure(figsize = (37.4/3, 10))
plt.imshow(MS99033_LED,alpha=0.6,cmap='Greys')
plt.imshow(np.flip(MS99033_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM, alpha = .6)
cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.0e')
cbar.ax.set_xlabel('Bz (T)', fontsize=12)

cbar.update_ticks()
plt.xticks([])
plt.yticks([])
# plt.savefig('../figure/MS99033_Bz_plot_with_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure(figsize = (37.4/3, 10))
plt.imshow(np.flip(MS99033_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM)

# cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.0e')
# cbar.ax.set_xlabel('Bz (T)', fontsize=12)

# cbar.update_ticks()
plt.xticks([])
plt.yticks([])
plt.box(False)
# plt.savefig('../figure/MS99033_Bz_plot_no_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```

## Duluth Complex anorthosite 696

```python
DC_696_Bz = loadmat('../data/QDM/DC_696/Bz_uc0.mat')
DC_696_Bz = DC_696_Bz['Bz'].reshape(300,374)

DC_696_LED = Image.open('../data/QDM/DC_696/LED.jpg').resize((374,300)).convert('LA')
```

```python
vmin_QDM=-.5e-4
vmax_QDM=.5e-4

plt.figure(figsize = (37.4/3, 10))
plt.imshow(DC_696_LED,alpha=0.6,cmap='Greys')
plt.imshow(np.flip(DC_696_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM, alpha = .6)
cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.0e')
cbar.ax.set_xlabel('Bz (T)', fontsize=12)

cbar.update_ticks()
plt.xticks([])
plt.yticks([])
plt.savefig('../figure/DC_696_Bz_plot_with_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure(figsize = (37.4/3, 10))
plt.imshow(np.flip(DC_696_Bz,0),cmap='seismic',vmin=vmin_QDM,vmax=vmax_QDM)

# cbar = plt.colorbar(fraction=0.05,shrink=0.85, format='%.0e')
# cbar.ax.set_xlabel('Bz (T)', fontsize=12)

# cbar.update_ticks()
plt.xticks([])
plt.yticks([])
plt.box(False)
plt.savefig('../figure/DC_696_Bz_plot_no_scale.png', bbox_inches='tight', dpi=300)
plt.show()
```
